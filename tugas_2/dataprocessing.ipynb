```python
import pandas as pd
import numpy as np
import random
from faker import Faker
```

```python
# Create a Faker instance
fake = Faker()

# Generate 20 random dealer names and IDs
dealer_data = []
for i in range(1, 21):
    dealer_id = i
    dealer_name = fake.company()
    dealer_data.append({'dealer_id': dealer_id, 'nama_dealer': dealer_name})

# Create a DataFrame from the dealer data
dealer = pd.DataFrame(dealer_data)
dealer
```

```python
def generate_transaction_id(iteration):
    return f"{iteration+1}"

# Set the number of iterations
num_iterations = 2000


# Generate random car IDs and city IDs
car_ids = np.random.randint(1, 51, size=num_iterations)
dealer_ids = np.random.randint(1, 21, size=num_iterations)
city_ids = np.random.randint(1, 16, size=num_iterations)

# Create an empty DataFrame
transaction = pd.DataFrame({'transaksi_id':[i+1 for i in range(num_iterations)],'dealer_id': dealer_ids,'product_id': car_ids, 'kota_id': city_ids})

transaction
```

```python
product.to_csv("data/product.csv",index=False)
city.to_csv("data/city.csv",index=False)
dealer.to_csv("data/dealer.csv",index=False)
transaction.to_csv("data/transaction.csv",index=False)
```
